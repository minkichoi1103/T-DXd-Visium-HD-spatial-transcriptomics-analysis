```python
import scanpy as sc 
import pandas as pd
import numpy as np
```

```python
## The tissue cores were initially segmented using `curvsplit`, as described in the Methods section.

## However, `curvsplit` may over-segment individual tissue cores into multiple small fragments,
## and some fragments may not contain any tumor components.

## This notebook demonstrates the process of visually inspecting the tissue images
## and manually merging fragments when appropriate.
```

```python
def transform_batch(numbers, category_prefix="B01"):
    return [f"{category_prefix}_{i}" for i in numbers]
```

```python
sp_path = '../data/h5ad_pbpk_55um'
```

```python
#1. B01
slide_id = 'B01'
adata = sc.read_h5ad(f'{sp_path}/sp_grid_P-1S-R-3K-{slide_id}-54-D_mf_pbpk_her2.h5ad')
adata = adata[adata.obs['subcluster'].isin([2, 3, 4, 5, 6, 9, 10 ])].copy()
adata.obs['batch'] = adata.obs['subcluster'].copy()
```

```python
adata.obs['batch'] = np.where(adata.obs['batch'].isin([3]), 2, adata.obs['batch'])
adata.obs['batch'] = adata.obs['batch'].astype('category')

adata.obs['batch'] = np.where(adata.obs['batch'].isin([5]), 4, adata.obs['batch'])
adata.obs['batch'] = adata.obs['batch'].astype('category')
```

```python
sc.pl.spatial(adata, color = 'batch', size = 7)
```

```python
nums = adata.obs['batch'].unique().tolist()
batch = transform_batch(nums, category_prefix = slide_id)
match_dict = dict(zip(nums, batch))
adata.obs['batch'] = adata.obs['batch'].astype(int).map(match_dict)
adata.obs['batch'] = adata.obs['batch'].astype('category')
```

```python
sc.pl.spatial(adata, color = 'batch', size = 7)
```

```python
adata.write_h5ad(f'../data/h5ad_batch_55um/{slide_id}.h5ad')
```
